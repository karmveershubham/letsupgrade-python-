```python
marks = int(input("Enter the marks "))    
if marks >= 95 and marks<101:   
    print('Grade is --> A+')
elif marks >=80 and marks <95:
    print('Grade is -->  A ')
elif marks >=70 and marks <80:
    print('Grade is -->  B+')
elif marks >=60 and marks <70:
    print('Grade is -->  B ')
elif marks >=50 and marks <60:
    print('Grade is -->  C+ ')    
elif marks >=40 and marks <50:
    print('Grade is -->  C ')
elif marks >=30 and marks <40:
    print('Grade is -->  D ')
else:
    print("Failed")
```

```python
#Question 2
#Use For Loop to Print Prime Numbers in between 1 to 1000
```

```python
for num in range(1,1000):
    prime=True
    for i in range(2,num):
        if num%i == 0:
            prime=False
    if prime:
         print(num)
```

```python
#Question 4
#Write a program to Print X Prime Numbers using While Loop starting from 0, and take the INput of X from the User
```

```python
x=int(input('enter value of  x'))
for a in range(2,x+1):
    prime= True
    for i in range (2,a):
        if a%i==0:
            prime=False
    if prime:
        print (a, end=" ")
```
